# MatplotLib

```python
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics.scorer import r2_score
import matplotlib.pyplot as plt
import numpy as np
```

```python
boston = load_boston()
print("Boston data: \n{}".format(boston.data[:4]))
print("Boston data size: \n{}".format(boston.data.shape))
print("Boston target: \n{}".format(boston.target[:40]))
print("Boston feature names: \n{}".format(boston.feature_names))
# print("Decsription: \n{}".format(boston.DESCR))
```

```python
X, y = boston.data, boston.target
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=42)

score_arrange = []
neighbor_setting = range(1,12)
for n_neighbor in neighbor_setting:
    knn_reg = KNeighborsRegressor(n_neighbors=n_neighbor).fit(X_train, y_train)
    print("N-neghbors: {}; Score: {:.2f}".format(n_neighbor, knn_reg.score(X_test, y_test)))
    score_arrange.append(knn_reg.score(X_test, y_test))
```

```python
plt.figure(figsize=(10, 6))
plt.plot(neighbor_setting, score_arrange, linewidth=3, marker='o', c='red', 
         markersize = 12, markerfacecolor='orange');
plt.grid(True)
plt.xlabel("N_neighbor", size=15);
plt.ylabel("Accuracy", size=15);
```

```python
# calculate R2 score
predict = knn_reg.predict(X_test)
print(r2_score(y_test, predict))
```

```python
# Output all data feature graps
rows = boston.data.shape[1]
columns = 1
fig, axs = plt.subplots(rows, columns, figsize=(20,100))
i = 0
for feature in boston.feature_names:    
    axs[i].plot(boston.data[:, i], '-', linewidth=3)
    axs[i].set_xlabel("Sample", size=15)
    axs[i].set_ylabel(feature, size=15)
    i += 1
```
